```python
import random
#keeps track of lives left
def update(progress):
    if progress == 0:
        print("You have 10 lives left")
        return
    for i in range(0, progress): 
        if i == 0 and progress == 1:
            print("You have 9 lives left.")
        if i == 1 and progress == 2:
            print("You have 8 lives left.")
        elif i == 2 and progress == 3:
            print("You have 7 lives left.")
        elif i == 3 and progress == 4:
            print("You have 6 lives left.")
        elif i == 4 and progress == 5:
            print("You have 5 lives left.")
        elif i == 5 and progress == 6:
            print("You have 4 lives left.")
        elif i == 6 and progress == 7:
            print("You have 3 lives left.")
        elif i == 7 and progress == 8:
            print("You have 2 lives left.")
        elif i == 8 and progress == 9:
            print("You have 1 lives left.")
        elif i == 9 and progress == 10:
            print("You have 0 lives left.")
    return

words = ["WHDSPQZ", "XHO", "TTDBFRT", "QQJYF", "ENQD", "DNPK", "CNTR", "EHWHOD", "TSVMOHOF", "XNOCFQGTM"]
word = random.choice(words)
#decrypts the word chosen 
def decrypt(ciphertext):
    ciphertext == word
    plaintext = ""
    i = 0
    if i < len(ciphertext):
        if i % 2:
            cipher_ascii = ord(ciphertext[i])
            plain_ascii = cipher_ascii - 1
            plaintext += chr(plain_ascii)
            i += 1
        elif i % 2 == 0:
            cipher_ascii = ord(ciphertext[i])
            plain_ascii = cipher_ascii + 1
            plaintext += chr(plain_ascii)
            i += 1
    else:
        print(str(plaintext))

word_progress = ["-"]*len(word)

guessed = []

incorrect = 0
#progress report and updating the lives count
while incorrect < 10 and "-" in word_progress:
    print()
    update(incorrect)
    print(" ".join(word_progress))
    print("You have used these letters: ")
    print(" ".join(guessed))
    print()
    
    guess = "0" #check for letter
    if not guess.isalpha():
        guess = input("Guess a letter: ").upper()
 #convert all guesses to uppercase       
    if guess == word.upper():
        break #word was guessed so you win
    elif len(guess) > 1: #guess more than one letter and incorrect
        print("That is not the correct word.")
        incorrect += 1
    elif guess in guessed:
        print("You have already guessed that letter.")
    elif guess not in word.upper():
        print("Your letter, " + guess + " is not in the word.")
        incorrect += 1
        guessed.append(guess)
    else: #add back to letters guessed
        guessed.append(guess)
        for i in range(len(word)):
            if word_progress[i] == "-" and word[i].upper() == guess:
                word_progress[i] = word[i]

                             
print()
if incorrect == 10:
    print("You lost, sorry. The word was: " + word)
    print("You'll have to guess the word correctly to decode the message")
else:
    if "-" in word_progress: #if you won by guessing the word
        print("You win!")
        print("This was the decrypted message!")
        print("The decoded message says: " + decrypt(ciphertext))
        for i in range(len(word)):
            word_progress[i] = word[i]
    else:  #if you won by guessing the letters
        print("You win!")
        print("This was the decrypted message!")
        print("The decoded message says: " + decrypt(ciphertext))
        
```
